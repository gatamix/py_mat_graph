```python
%matplotlib inline
from matplotlib import rcParams
import matplotlib.pyplot as plt
params={'font.family' : "Noto Sans CJK JP",'font.size':14,'font.weight':"regular"}
rcParams.update(params)


fig, ax = plt.subplots()
ax.plot([1, 2, 3],label = 'test1')

plt.title('Noto Sans CJK JP',fontweight = "regular",fontsize = 20)
plt.xlabel('X軸ラベル' ,fontweight = "regular" ,fontsize = 20)
plt.ylabel('Y軸ラベル' ,fontweight = "regular" ,fontsize = 20)


ax.legend()
plt.show()
```

```python
%matplotlib inline
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
from matplotlib import rcParams

my_font = fm.FontProperties(fname='/System/Library/Fonts/ヒラギノ角ゴシック W5.ttc')
plt.rcParams.update({'font.size': 14})


fig, ax = plt.subplots()
ax.plot([1, 2, 3], label = 'test')

plt.xlabel('X label' ,fontsize = 20)
plt.ylabel('Y label' ,fontsize = 20)

ax.legend(prop=my_font)

plt.title('ヒラギノ角ゴシック', fontproperties=my_font,fontsize = 20)
plt.xlabel('X軸ラベル', fontproperties=my_font,fontsize = 20)
plt.ylabel('y軸ラベル', fontproperties=my_font,fontsize = 20)
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontproperties(my_font)

for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontproperties(my_font)

plt.show()
#plt.savefig("font_test.svg")
```
