```python
from matplotlib import rcParams
import matplotlib.pyplot as plt
params={'font.family' : "Noto Sans CJK JP",'font.size':14,'font.weight':"regular"}
rcParams.update(params)


fig, ax = plt.subplots()
ax.plot([1, 2, 3],label = 'test1')
plt.xlabel('X軸ラベル' ,fontweight = "regular" ,fontsize = 20)
plt.ylabel('Y軸ラベル' ,fontweight = "regular" ,fontsize = 20)


ax.legend()
plt.show()
```

```python
from matplotlib import rcParams
import matplotlib.pyplot as plt
params={'font.family' : "Noto Sans CJK JP",'font.size':14,'font.weight':"regular"}
rcParams.update(params)


fig, ax = plt.subplots()
ax.plot([1, 2, 3],[1, 2, 3],'bo-',label = 'test1')
ax.plot([1, 2, 3],[3, 6, 9],'ro--',label = 'test2')
plt.xlabel('X軸ラベル' ,fontweight = "regular" ,fontsize = 20)
plt.ylabel('Y軸ラベル' ,fontweight = "regular" ,fontsize = 20)


ax.legend()
plt.show()
```
