```python
%matplotlib inline
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
from matplotlib import rcParams

my_font = fm.FontProperties(fname=r'C:\Users\font\NotoSansCJKjp-Regular.otf')
plt.rcParams.update({'font.size': 14})


fig, ax = plt.subplots()
ax.plot([1, 2, 3], label = 'test')

plt.xlabel('X label' ,fontsize = 20)
plt.ylabel('Y label' ,fontsize = 20)

ax.legend(prop=my_font)

plt.title('Noto Sans CJK jp', fontproperties=my_font,fontsize = 20)
plt.xlabel('X軸ラベル', fontproperties=my_font,fontsize = 20)
plt.ylabel('y軸ラベル', fontproperties=my_font,fontsize = 20)
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontproperties(my_font)

for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontproperties(my_font)

plt.show()
#plt.savefig("font_test.svg")
```
